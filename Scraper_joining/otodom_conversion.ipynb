```python
import pandas as pd
import numpy as np
```

```python
otodom_data = pd.read_csv('otodom_data.csv', sep='|')
otodom_data
```

## Data conversion and cleaning

```python
otodom_data['price_per_m2'] = round(otodom_data['price[PLN]'] / otodom_data['area[m^2]'], 2)
otodom_data
```

```python
otodom_data.info()
```

```python
otodom_data['addition'].value_counts()
```

```python
otodom_data['addition'] = otodom_data['addition'].str.replace(' ', '')
dummies = otodom_data['addition'].str.get_dummies(sep=',')
otodom_data = pd.concat([otodom_data, dummies], axis=1).drop('addition', axis=1)
otodom_data
```

```python
otodom_data.rename(columns={'buiiding_type': 'building_type'}, inplace=True)
otodom_data['building_type'].value_counts()
```

```python
# Floor column
otodom_data['floor'] = otodom_data['floor'].str.replace('parter', '0')

# market column
otodom_data['market'] = otodom_data['market'].str.replace('wtórny', 'secondary')
otodom_data['market'] = otodom_data['market'].str.replace('pierwotny', 'primary')

# building_type column
otodom_data['building_type'] = otodom_data['building_type'].str.replace('blok', 'Mieszkanie')
otodom_data['building_type'] = otodom_data['building_type'].str.replace('kamienica', 'Mieszkanie')
otodom_data['building_type'] = otodom_data['building_type'].str.replace('apartamentowiec', 'Mieszkanie')
# replace wolnostojący with dom wolnostojący
otodom_data['building_type'] = otodom_data['building_type'].str.replace('dom wolnostojący', 'wolnostojący')
```

```python
otodom_data['building_type'].value_counts()
```

```python
otodom_data
```

```python
otodom_data['parking'] = otodom_data['parking'].apply(lambda x: 1 if x is not np.nan else 0)
otodom_data
```

```python
otodom_data['elevator'] = otodom_data['elevator'].apply(lambda x: 1 if x == 'tak' else 0)
otodom_data
```

```python
otodom_data['address'] = otodom_data['address'].str.replace(', pomorskie', '')
otodom_data['address'] = otodom_data['address'].str.replace(', ', ',')
otodom_data
```

```python
def add_null_prefix(address):
    if str(address).count(',') == 1:
        return 'null,' + str(address)
    return str(address)

otodom_data['address'] = otodom_data['address'].apply(add_null_prefix)

otodom_data
```

```python
otodom_data = otodom_data[otodom_data['address'].str.count(',') == 2]
otodom_data
```

```python
otodom_data[['street', 'district', 'city']] = otodom_data['address'].str.split(',', expand=True)
otodom_data
```

```python
otodom_data.drop('address', axis=1, inplace=True)
otodom_data
```

```python
otodom_data.to_csv('otodom_data_cleaned.csv', sep='|', index=False)
```
